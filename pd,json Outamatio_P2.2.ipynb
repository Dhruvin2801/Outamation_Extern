```python
# 2. Fill missing values (numeric ‚Üí mean, categorical ‚Üí mode)
for col in df.columns:
    if df[col].dtype == 'object':  # categorical
        df[col].fillna(df[col].mode()[0], inplace=True)
    else:  # numeric
        df[col].fillna(df[col].mean(), inplace=True)

# 3. Drop duplicates
df.drop_duplicates(inplace=True)

# 4. Save cleaned dataset
df.to_csv('bank_loan_cleaned_simple.csv', index=False)
```

# JSON

```python
import json
import pandas as pd
from pandas import json_normalize

# ---- Load Nested JSON ----
with open("sample.json") as f:
    data = json.load(f)

# ---- Flatten Orders (customer + order details) ----
df_orders = json_normalize(
    data,
    record_path=['orders'],                     # go inside orders
    meta=['customer_id', 'name', 'email']       # keep customer info
)

print("=== Orders Table ===")
print(df_orders)

# ---- Flatten Items (one row per product) ----
orders_list = []
for cust in data:
    for order in cust["orders"]:
        for item in order["items"]:
            orders_list.append({
                "customer_id": cust["customer_id"],
                "name": cust["name"],
                "email": cust["email"],
                "order_id": order["order_id"],
                "date": order["date"],
                "payment_method": order["payment"]["method"],
                "payment_status": order["payment"]["status"],
                "product": item["product"],
                "price": item["price"],
                "quantity": item["quantity"]
            })

df_items = pd.DataFrame(orders_list)

print("\n=== Items Table ===")
print(df_items)

# ---- Export Flattened Data Back to JSON ----
df_orders.to_json("flattened_orders.json", orient="records", indent=4)
df_items.to_json("flattened_items.json", orient="records", indent=4)

print("\nFlattened JSON files saved: flattened_orders.json & flattened_items.json")
```

# **TEXT_SAMPLE**

```python
import re
import html
from bs4 import BeautifulSoup
import contractions
from nltk.corpus import stopwords
import nltk

# --- Download stopwords once (comment out after first run) ---
nltk.download('stopwords')

# --- Load messy text file ---
try:
    with open("/content/text_sample.txt", "r", encoding="utf-8", errors="ignore") as f:
        text = f.read()
except FileNotFoundError:
    print("Error: messy_text.txt not found. Please upload the file.")
    text = "" # Set empty string if file not found


# --- Decode HTML entities & remove HTML tags ---
text = html.unescape(text)
text = BeautifulSoup(text, "html.parser").get_text(separator=" ")

# --- Expand contractions (can't -> cannot) ---
text = contractions.fix(text)

# --- Normalize case ---
text = text.lower()

# --- Standardize prices (convert formats like 29,99$ -> 29.99 usd) ---
text = re.sub(r'\$?(\d+)[,](\d+)\$?', r'\1.\2 usd', text)
text = re.sub(r'\$([\d.]+)', r'\1 usd', text)

# --- Remove emails and phone numbers ---
text = re.sub(r'\S+@\S+\.\S+', ' ', text)
text = re.sub(r'\(?\d{3}\)?[-.\s]?\d{3}[-.\s]?\d{4}', ' ', text)

# --- Remove random numeric sequences (optional) ---
text = re.sub(r'\b\d{4,}\b', ' ', text)

# --- Remove special characters & repeated punctuation ---
text = re.sub(r'[^\w\s]', ' ', text)           # keep words and spaces
text = re.sub(r'\s+', ' ', text)               # collapse multiple spaces

# --- Remove stopwords ---
# Make sure to download stopwords if you haven't already by uncommenting and running the nltk.download('stopwords') line
try:
    stop_words = set(stopwords.words('english'))
    words = [w for w in text.split() if w not in stop_words]
    text = ' '.join(words)
except LookupError:
    print("NLTK stopwords not found. Please uncomment and run the nltk.download('stopwords') line.")
    words = text.split()
    text = ' '.join(words)


# --- Save cleaned text ---
with open("cleaned_text.txt", "w", encoding="utf-8") as f:
    f.write(text)

print("Cleaned text saved to cleaned_text.txt")
```

# IMAGE PREPROCESSING

```python
import cv2
from google.colab.patches import cv2_imshow

# --- 1. Load Image ---
img = cv2.imread('/content/noisy_image_sample.jpg')

print("Original Image:")
cv2_imshow(img)

# --- 2. Noise Reduction Filters ---
gaussian = cv2.GaussianBlur(img, (5, 5), 0)
median = cv2.medianBlur(gaussian, 5)
bilateral = cv2.bilateralFilter(median, 9, 75, 75)

print("\nAfter Noise Reduction (Bilateral):")
cv2_imshow(bilateral)

# --- 3. Non-Local Means Denoising ---
denoised = cv2.fastNlMeansDenoisingColored(bilateral, None, 10, 10, 7, 21)

print("\nAfter Non-Local Means Denoising:")
cv2_imshow(denoised)

# --- 4. Convert to Grayscale ---
gray = cv2.cvtColor(denoised, cv2.COLOR_BGR2GRAY)

print("\nGrayscale Image:")
cv2_imshow(gray)

# --- 5. Contrast Enhancement with CLAHE ---
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
clahe_img = clahe.apply(gray)

print("\nAfter CLAHE (Contrast Enhanced):")
cv2_imshow(clahe_img)

# --- 6. Adaptive Thresholding for OCR Readiness ---
thresh_adaptive = cv2.adaptiveThreshold(clahe_img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                        cv2.THRESH_BINARY, 11, 2)

print("\nAfter Adaptive Thresholding:")
cv2_imshow(thresh_adaptive)

# --- 7. Save the Cleaned Image ---
cv2.imwrite("cleaned_image.jpg", thresh_adaptive)
print("\n‚úÖ Cleaned image saved as cleaned_image.jpg")
```
